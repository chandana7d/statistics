```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import matplotlib.pyplot as plt
import seaborn as sns
import torch
```

To compare two or more distributions, several statistical methods and visual tools can be used depending on the nature of the distributions (discrete vs. continuous) and what aspects you want to compare (shape, mean, variance, etc.). Below are several ways to compare distributions.



### 1. Histogram Method


```python



# Set the number of samples
num_samples = 100

# Generate random data for two normal distributions using PyTorch
data1 = torch.normal(mean=0, std=1, size=(num_samples,))  # Normal Distribution 1
data2 = torch.normal(mean=1, std=1.5, size=(num_samples,))  # Normal Distribution 2

# Plot histograms
plt.hist(data1.numpy(), bins=30, alpha=0.5, label='Distribution 1')
plt.hist(data2.numpy(), bins=30, alpha=0.5, label='Distribution 2')

# Add axis labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.title('Histogram Comparison')

# Show the plot
plt.show()
```

### 2. Box plot Method


```python

# Set number of samples
num_samples = 1000

# Generate random data for two normal distributions using PyTorch
data1 = torch.normal(mean=0, std=1, size=(num_samples,))
data2 = torch.normal(mean=1, std=1.5, size=(num_samples,))

data = [data1.numpy(), data2.numpy()]  # Convert tensors to NumPy arrays for Seaborn

# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(8, 3))

# Box Plot
sns.boxplot(data=data, ax=axes[0])
axes[0].set_title('Box Plot Comparison')

# Violin Plot
sns.violinplot(data=data, ax=axes[1])
axes[1].set_title('Violin Plot Comparison')

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import scipy.stats as stats

# Set number of samples
num_samples = 1000

# Generate random data for two normal distributions using PyTorch
data1 = torch.normal(mean=0, std=1, size=(num_samples,)).numpy()  # Normal Distribution 1
data2 = torch.normal(mean=1, std=1.5, size=(num_samples,)).numpy()  # Normal Distribution 2

data = [data1, data2]  # Convert tensors to NumPy arrays

# Create a figure with three subplots: Box, Violin, and QQ Plot
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Box Plot
sns.boxplot(data=data, ax=axes[0])
axes[0].set_title('Box Plot Comparison')

# Violin Plot
sns.violinplot(data=data, ax=axes[1])
axes[1].set_title('Violin Plot Comparison')

# QQ Plot for one of the distributions (data1)
stats.probplot(data1, dist="norm", plot=axes[2])
axes[2].set_title('QQ Plot (Data1 vs Normal)')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
